1. Write a Python program to find sum of elements in list?

```python
l = input('Enter numbers separated by space: ')
l = l.split()
total = sum([float(i) for i in l])
print(total)
```

2. Write a Python program to Multiply all numbers in the list?

```python
l = input('Enter numbers separated by space: ')
l = l.split()
l = [float(i) for i in l]
prod = 1
for i in l:
    prod *= i
print(prod)
```

3. Write a Python program to find smallest number in a list?

```python
l = input('Enter numbers separated by space: ')
l = l.split()
minimum = min([float(i) for i in l])
print(minimum)
```

4. Write a Python program to find largest number in a list?

```python
l = input('Enter numbers separated by space: ')
l = l.split()
maximum = max([float(i) for i in l])
print(maximum)
```

5. Write a Python program to find second largest number in a list?

```python
l = input('Enter numbers separated by space: ')
l = l.split()
l = [float(i) for i in l]
sorted_l = sorted(l, reverse=True)
second_largest = sorted_l[1]
print(second_largest)
```

6. Write a Python program to find N largest elements from a list?

```python
l = input('Enter numbers separated by space: ')
n = int(input('Enter N to find N largest numbers: '))
l = l.split()
l = [float(i) for i in l]
sorted_l = sorted(l, reverse=True)
n_largest = sorted_l[: n]
print(n_largest)
```

7. Write a Python program to print even numbers in a list?

```python
l = input('Enter numbers separated by space: ')
l = l.split()
even = [float(i) for i in l if float(i) % 2 == 0]
print(even)
```

8. Write a Python program to print odd numbers in a List?

```python
l = input('Enter numbers separated by space: ')
l = l.split()
odd = [float(i) for i in l if float(i) % 2 == 1]
print(odd)
```

9. Write a Python program to Remove empty List from List?

```python
l = [1, 2, [], [1, 2], [], [4, 5, 7], 'a', 3.65, 'str']
new_l = []
for i in l:
    if type(i) == list:
        if len(i) != 0:
            new_l.append(i)
    else:
        new_l.append(i)
        
print(new_l)
```

10. Write a Python program to Cloning or Copying a list?

```python
l = [1, 2, 3]
l_copy = l.copy()
print(l)
print(l_copy)
```

11. Write a Python program to Count occurrences of an element in a list?

```python
l = input('Enter numbers separated by space: ')
to_find = input('Enter element to find the count of: ')
l = l.split()
count = l.count(to_find)
print(count)
```
